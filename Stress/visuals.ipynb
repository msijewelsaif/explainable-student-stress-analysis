```python
import matplotlib.pyplot as plt
import pandas as pd

# Your data
data = {
    "Model": ["Random Forest", "MLP Neural Network", "SVM", "Logistic Regression", "Gradient Boosting"],
    "Accuracy": [0.918182, 0.890909, 0.890909, 0.881818, 0.863636],
    "Precision": [0.918647, 0.890956, 0.892235, 0.882194, 0.865773],
    "Recall": [0.918182, 0.890909, 0.890909, 0.881818, 0.863636],
    "F1-score": [0.918220, 0.890906, 0.890737, 0.881437, 0.864337]
}

df = pd.DataFrame(data)

# Plot
metrics = ["Accuracy", "Precision", "Recall", "F1-score"]
x = range(len(df["Model"]))
bar_width = 0.2

plt.figure(figsize=(10,6))

for i, metric in enumerate(metrics):
    plt.bar([p + i*bar_width for p in x], df[metric], width=bar_width, label=metric)

# Formatting
plt.xticks([p + 1.5*bar_width for p in x], df["Model"], rotation=20, ha="right")
plt.ylabel("Score")
plt.title("Model Performance Comparison")
plt.legend()
plt.tight_layout()
plt.show()
```
